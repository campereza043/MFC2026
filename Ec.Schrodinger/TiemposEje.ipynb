## PROBLEMA 1 PYTHON

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
enmallados = ['M=20', 'M=50', 'M=1000']
tiempos = [0.183, 0.615, 207]

# Configuración estética
plt.figure(figsize=(10, 6))
colores = ['#3498db', '#2ecc71', '#e74c3c'] # Azul, Verde, Rojo

# Crear gráfica de barras
bars = plt.bar(enmallados, tiempos, color=colores, edgecolor='black', alpha=0.8)

# Añadir los valores exactos encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval} s', 
             ha='center', va='bottom', fontweight='bold')

# Personalización de etiquetas y título
plt.title('Comparación de Tiempos de Ejecución por Enmallado (M)', fontsize=14, pad=20)
plt.xlabel('Número de puntos del enmallado (M)', fontsize=12)
plt.ylabel('Tiempo de ejecución (segundos)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Opcional: Si quieres que se vea mejor la diferencia entre los pequeños, 
# puedes descomentar la siguiente línea para usar escala logarítmica:
# plt.yscale('log')



plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
enmallados = ['M=20', 'M=50', 'M=1000']
tiempos = [0.183, 0.615, 207]

# Configuración de la figura
plt.figure(figsize=(10, 6))
colores = ['#5DADE2', '#58D68D', '#EC7063']

# Crear gráfica de barras
bars = plt.bar(enmallados, tiempos, color=colores, edgecolor='black', alpha=0.8)

# ACTIVAR ESCALA LOGARÍTMICA
plt.yscale('log')

# Añadir etiquetas de datos
# En escala logarítmica, el desplazamiento vertical (yval * 1.2) debe ser multiplicativo
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval * 1.1, f'{yval} s', 
             ha='center', va='bottom', fontweight='bold', fontsize=10)

# Personalización
plt.title('Tiempo de Ejecución vs Resolución de Enmallado (Escala Logarítmica)', fontsize=14)
plt.xlabel('Enmallado (M)', fontsize=12)
plt.ylabel('Tiempo (s) - Escala Log', fontsize=12)

# Añadir una rejilla más detallada para la escala logarítmica
plt.grid(True, which="both", ls="-", alpha=0.2, axis='y')



plt.tight_layout()
plt.show()
```
