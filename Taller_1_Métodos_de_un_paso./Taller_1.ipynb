```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
# parametros del problema
rb = 0.002
h = 1
t0 = 0
tf = 50

# vector de tiempo
t = np.arange(t0, tf + h, h)
N = len(t)

# condicion inicial
p0 = 0.01

# solucion exacta
p_exact = 1 - 0.99 * np.exp(-rb * t)
p_exact_50 = 1 - 0.99 * np.exp(-rb * 50)

print("Valor exacto en t=50:", p_exact_50)
```

```python
p_euler = np.zeros(N)
p_euler[0] = p0

start = time.time()

for n in range(N-1):
    p_euler[n+1] = p_euler[n] + rb * (1 - p_euler[n])

end = time.time()
time_euler = end - start

print("Euler explícito en t=50:", p_euler[-1])
print("Tiempo de ejecución Euler:", time_euler, "segundos")
```

```python
plt.figure()
plt.plot(t, p_exact, label="Solución exacta")
plt.plot(t, p_euler, "--", label="Euler explícito")
plt.xlabel("t")
plt.ylabel("p(t)")
plt.title("Método Euler Explícito")
plt.legend()
plt.grid()
plt.show()
```

```python
p_taylor = np.zeros(N)
p_taylor[0] = p0

start = time.time()

for n in range(N-1):
    p_taylor[n+1] = p_taylor[n] + 0.001998 * (1 - p_taylor[n])

end = time.time()
time_taylor = end - start

print("Taylor orden 2 en t=50:", p_taylor[-1])
print("Tiempo de ejecución Taylor:", time_taylor, "segundos")
```

```python
plt.figure()
plt.plot(t, p_exact, label="Solución exacta")
plt.plot(t, p_taylor, "--", label="Taylor orden 2")
plt.xlabel("t")
plt.ylabel("p(t)")
plt.title("Método Taylor Orden 2")
plt.legend()
plt.grid()
plt.show()
```

```python
p_trap = np.zeros(N)
p_trap[0] = p0

start = time.time()

for n in range(N-1):
    p_trap[n+1] = (0.999 * p_trap[n] + 0.002) / 1.001

end = time.time()
time_trap = end - start

print("Trapecio en t=50:", p_trap[-1])
print("Tiempo de ejecución Trapecio:", time_trap, "segundos")
```

```python
plt.figure()
plt.plot(t, p_exact, label="Solución exacta")
plt.plot(t, p_trap, "--", label="Trapecio")
plt.xlabel("t")
plt.ylabel("p(t)")
plt.title("Método del Trapecio")
plt.legend()
plt.grid()
plt.show()
```

```python
# errores absolutos en t=50
error_euler = abs(p_exact_50 - p_euler[-1])
error_taylor = abs(p_exact_50 - p_taylor[-1])
error_trap = abs(p_exact_50 - p_trap[-1])

print("=== ERRORES ABSOLUTOS ===")
print("Euler:", error_euler)
print("Taylor:", error_taylor)
print("Trapecio:", error_trap)

print("\n=== TIEMPOS DE EJECUCION ===")
print("Euler:", time_euler, "segundos")
print("Taylor:", time_taylor, "segundos")
print("Trapecio:", time_trap, "segundos")
```

```python
plt.figure(figsize=(8,5))

plt.plot(t, p_exact, label="Solución exacta", linewidth=3)
plt.plot(t, p_euler, "--", label="Euler")
plt.plot(t, p_taylor, "--", label="Taylor 2")
plt.plot(t, p_trap, "--", label="Trapecio")

plt.xlabel("t")
plt.ylabel("p(t)")
plt.title("Comparación de Métodos Numéricos")
plt.legend()
plt.grid()

plt.show()
```

```python
# Euler - maximo error hasta 4000 años

import time
import numpy as np

rb = 0.002
h = 1
Tmax = 4000

p_num = 0.01
max_error = 0
year_max_error = 0

start = time.time()

for t_actual in range(Tmax + 1):
    
    # solucion exacta
    p_exact = 1 - 0.99*np.exp(-rb*t_actual)
    
    if t_actual > 0:  # evitar division por cero
        error = abs(p_exact - p_num) / abs(p_exact)
        
        if error > max_error:
            max_error = error
            year_max_error = t_actual
    
    # paso Euler
    p_num = p_num + rb*(1 - p_num)

end = time.time()

print("=== EULER ===")
print("Error máximo:", max_error)
print("Año donde ocurre:", year_max_error)
print("Tiempo ejecución:", end-start, "segundos")
```

```python
# Taylor orden 2 - maximo error hasta 4000 años

rb = 0.002
h = 1
Tmax = 4000

p_num = 0.01
max_error = 0
year_max_error = 0

start = time.time()

for t_actual in range(Tmax + 1):
    
    p_exact = 1 - 0.99*np.exp(-rb*t_actual)
    
    if t_actual > 0:
        error = abs(p_exact - p_num) / abs(p_exact)
        
        if error > max_error:
            max_error = error
            year_max_error = t_actual
    
    # paso Taylor orden 2
    p_num = p_num + 0.001998*(1 - p_num)

end = time.time()

print("=== TAYLOR ORDEN 2 ===")
print("Error máximo:", max_error)
print("Año donde ocurre:", year_max_error)
print("Tiempo ejecución:", end-start, "segundos")
```

```python
# Trapecio - maximo error hasta 4000 años

rb = 0.002
h = 1
Tmax = 4000

p_num = 0.01
max_error = 0
year_max_error = 0

start = time.time()

for t_actual in range(Tmax + 1):
    
    p_exact = 1 - 0.99*np.exp(-rb*t_actual)
    
    if t_actual > 0:
        error = abs(p_exact - p_num) / abs(p_exact)
        
        if error > max_error:
            max_error = error
            year_max_error = t_actual
    
    # paso Trapecio
    p_num = (0.999*p_num + 0.002) / 1.001

end = time.time()

print("=== TRAPECIO ===")
print("Error máximo:", max_error)
print("Año donde ocurre:", year_max_error)
print("Tiempo ejecución:", end-start, "segundos")
```
