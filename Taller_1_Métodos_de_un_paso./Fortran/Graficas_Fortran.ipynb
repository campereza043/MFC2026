```python
import numpy as np
import matplotlib.pyplot as plt

# cargar datos
exact = np.loadtxt("exacta.dat")
euler = np.loadtxt("euler.dat")
taylor = np.loadtxt("taylor.dat")
trap = np.loadtxt("trapecio.dat")

# separar columnas
t_exact, p_exact = exact[:,0], exact[:,1]
t_euler, p_euler = euler[:,0], euler[:,1]
t_taylor, p_taylor = taylor[:,0], taylor[:,1]
t_trap, p_trap = trap[:,0], trap[:,1]

# grafica comparativa
plt.figure(figsize=(8,5))

plt.plot(t_exact, p_exact, label="Solucin exacta", linewidth=3)
plt.plot(t_euler, p_euler, "--", label="Euler explcito")
plt.plot(t_taylor, p_taylor, "--", label="Taylor orden 2")
plt.plot(t_trap, p_trap, "--", label="Trapecio")

plt.xlabel("Tiempo (aos)")
plt.ylabel("p(t)")
plt.title("Comparacin de mtodos numricos")
plt.legend()
plt.grid()

plt.show()
```
